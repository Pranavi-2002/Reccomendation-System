```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True)
```

```python
books_df = df.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
```

```python
books_df
```

```python
#Impute those NaNs with 0 values
books_df.fillna(0, inplace=True)
```

```python
books_df
```

```python
books_df.index=df['User.ID'].unique()
```

```python
books_df
```

```python
#Calculating cosine Similarity between Users
user_sim=1-pairwise_distances(books_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```
